# Example of how to use the ADSA package.

```python
import numpy as np
```

```python
%load_ext autoreload
%autoreload 2
```

```python
import adsa
```

```python
R0 = 10.0e-3 # 10 mm
ca = 150.0

(alphas, xs, zs) = adsa.solver.simulate_droplet_shape(R0, ca)
vol = adsa.analysis.calculate_volume(xs, zs, R0)
adsa.visualisation.plot_drop(xs, zs)

print(u"Contact angle: {:.4}°\nVolume: {:.4} uL".format(alphas[-1], vol*1e9))

print(f"{bottom_area} mm2".format(bottom_area=np.pi*xs[-1]**2)
```

```python
V = 10.0e-12 # 10 uL
ca = 150.0

(alphas, xs, zs) = adsa.solver.simulate_droplet_shape_for_volume(V, ca, R0_guess=20e-3, n_steps=200)
vol = adsa.analysis.calculate_volume(xs, zs, R0)
adsa.visualisation.plot_drop(xs, zs)

print(u"Contact angle: {:.4}°\nVolume: {:.4} uL".format(alphas[-1], vol*1e9))
```

```python
h = 5.0e-3 # 5 mm
ca = 150.0

(alphas, xs, zs) = adsa.solver.simulate_droplet_shape_for_height(h, ca, n_steps=200)
vol = adsa.analysis.calculate_volume(xs, zs, R0)
adsa.visualisation.plot_drop(xs, zs)

print(u"Contact angle: {:.4}°\nVolume: {:.4} uL".format(alphas[-1], vol*1e9))
```
